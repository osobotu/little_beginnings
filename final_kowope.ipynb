## Where it all started! 

```python
# load the data

import pandas as pd
import numpy as np

train = pd.read_csv("C:\\Users\\USER\\Desktop\\Qualification_Hack\\Train.csv")
test = pd.read_csv("C:\\Users\\USER\\Desktop\\Qualification_Hack\\Test.csv")
```

```python
train = train.dropna(axis=0, thresh=15)
train.shape
train = train.fillna(-999)
test = test.fillna(-999)
```

```python
# extract features and target
X = train.drop(["Applicant_ID", "default_status"], axis=1)
y = train['default_status'].replace(['no', 'yes'], [0, 1])

test_X = test.drop(["Applicant_ID"], axis=1)
```

```python
X['form_field47'] = X['form_field47'].replace(['charge', 'lending'], [0, 1])
test_X['form_field47'] = test_X['form_field47'].replace(['charge', 'lending'], [0, 1])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
X = pd.DataFrame(X)
test_X = scaler.transform(test_X)
test_X = pd.DataFrame(test_X)
```

## From here on I started trying out different models

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score, StratifiedKFold

lgbm = LGBMClassifier()

cv = StratifiedKFold(n_splits=35)
n_scores = cross_val_score(lgbm, X, y, scoring='roc_auc', cv=cv, n_jobs=-1,error_score='raise')
print("ROC_AUC: %.3f (%.3f)" % (np.mean(n_scores), np.std(n_scores)))
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier()

cv = StratifiedKFold(n_splits=10,shuffle=True, random_state=1)
n_scores = cross_val_score(forest, X, y, scoring='roc_auc', cv=cv, n_jobs=-1,error_score='raise')
print("ROC_AUC: %.3f (%.3f)" % (np.mean(n_scores), np.std(n_scores)))
```

```python
from rgf.sklearn import RGFClassifier

rgf = RGFClassifier()

cv = StratifiedKFold(n_splits=10,shuffle=True, random_state=1)
n_scores = cross_val_score(rgf, X, y, scoring='roc_auc', cv=cv, n_jobs=-1,error_score='raise')
print("ROC_AUC: %.3f (%.3f)" % (np.mean(n_scores), np.std(n_scores)))
```

```python
from catboost import CatBoostClassifier

cat = CatBoostClassifier(bagging_temperature=0.5434030676903125,
                                depth=7, iterations=502, l2_leaf_reg=25,
                                learning_rate=0.04447541043186938,
                                random_strength=3.245977736555784e-09,
                                scale_pos_weight=0.5750700246521092,
                       #cat_features=cat_features
                        )

cv = StratifiedKFold(n_splits=10,shuffle=True, random_state=1)
n_scores = cross_val_score(cat, X, y, scoring='roc_auc', cv=cv, n_jobs=-1,error_score='raise')
#roc_auc_score(y_val, catpred)
print("ROC_AUC: %.3f (%.3f)" % (np.mean(n_scores), np.std(n_scores)))
```

```python
from sklearn.ensemble import VotingClassifier

estimators = []

estimators.append(('lgbm', lgbm))
estimators.append(('cat', cat))
estimators.append(('forest', forest))
estimators.append(('rgf', rgf))


vot_soft = VotingClassifier(estimators=estimators, n_jobs=-1, voting='soft',weights=[3,1,1,2] )
vot_soft.fit(X, y)
```

```python
cv = StratifiedKFold(n_splits=10,shuffle=True, random_state=1)
n_scores = cross_val_score(vot_soft, X, y, scoring='roc_auc', cv=cv, n_jobs=-1,error_score='raise')
#roc_auc_score(y_val, catpred)
print("ROC_AUC: %.3f (%.3f)" % (np.mean(n_scores), np.std(n_scores)))
```

```python
finalpred = vot_soft.predict_proba(test_X)[:, 1]
pd.DataFrame({"Applicant_ID": test["Applicant_ID"], "default_status": finalpred}).to_csv("last_days3.csv", index = False)
```

## Hyperparameter tuning

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from pprint import pprint

# make list of parameters to search over.
n_estimators = [int(x) for x in np.linspace(start=20, stop=200, num=5)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(1, 45, num=3)]
min_samples_split = [5, 10]

# create random grid
random_grid = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split}

rf_random =  RandomizedSearchCV(estimator=forest, param_distributions=random_grid, n_iter=10, cv=10, verbose=2, random_state=42,
                               n_jobs=-1, scoring='roc_auc')

rf_random.fit(X_train, y_train)
```

```python
forest1 = rf_random.best_estimator_

cv = StratifiedKFold(n_splits=10,shuffle=True, random_state=1)
n_scores = cross_val_score(forest1, X, y, scoring='roc_auc', cv=cv, n_jobs=-1,error_score='raise')
print("ROC_AUC: %.3f (%.3f)" % (np.mean(n_scores), np.std(n_scores)))
```
